```
%matplotlib inline
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime as dt
import csv
from tqdm import tqdm_notebook as tqdm
import codecs
```

```
with codecs.open('/content/drive/My Drive/Colab Notebooks/どうでしょう新作uuid.csv', 'r', 'utf-8', 'ignore') as f:
  lines = f.readlines()

Wed4_uuid = []
Wed5_uuid = []
Wed6_uuid = []
Wed7_uuid = []
Wed8_uuid = []
Wed9_uuid = []
Wed10_uuid = []
Wed11_uuid = []
for line in lines[2:]:
  items = line.rstrip().split(",")
  if items[0]!= "":
    Wed4_uuid.append(items[0])
  if items[2]!= "": 
    Wed5_uuid.append(items[2])
  if items[4]!= "":
    Wed6_uuid.append(items[4])
  if items[6]!= "":
    Wed7_uuid.append(items[6])
  if items[8]!= "":
    Wed8_uuid.append(items[8])
  if items[10]!= "":
    Wed9_uuid.append(items[10])
  if items[12]!= "":
    Wed10_uuid.append(items[12])
  if items[14]!= "":
    Wed11_uuid.append(items[14])
```
